```python
from matplotlib.pyplot import imshow

from models.models import SegmentationModule, build_encoder, build_decoder
from src.eval import segment_image
from utils.constants import DEVICE
import cv2 as cv
```

```python
# path to image on which to run wall segmentation
path_image = 'img1.jpg'

# Model weights (encoder and decoder)
# For wall segmentation use weights from Drive (see model_weights/README.md):
#   - "Transfer learning - entire decoder": transfer_encoder.pth, transfer_decoder.pth
#   - "Without transfer learning": best_encoder_epoch_19.pth, best_decoder_epoch_19.pth
weights_encoder = 'model_weights/transfer_encoder.pth'
weights_decoder = 'model_weights/transfer_decoder.pth'
```

```python
net_encoder = build_encoder(weights_encoder)
net_decoder = build_decoder(weights_decoder)

segmentation_module = SegmentationModule(net_encoder, net_decoder)
segmentation_module = segmentation_module.to(DEVICE).eval()
```

```python
# Predict wall on input image and return segmentation mask
# max_size=512 keeps inference within ~6GB VRAM; use 768 or None if you have more GPU memory
segmentation_mask = segment_image(segmentation_module, path_image, max_size=512)
```
